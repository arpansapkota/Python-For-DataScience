# How Python works?


1. The interpreter reads a Python expression or statement, also called the source code, and verifies that it is well formed.As soon as the interpreter encounters error, it halts translation with an error message.

2. If a  Python  expression  is  well  formed,  the  interpreter  then  translates  it  to  an equivalent form in a low-level language called byte code. When the interpreter runs a script, it completely translates it to byte code.

3. This byte code is next sent to another software component, called the Python virtual  machine  (PVM),  where  it  is  executed.  If  another  error  occurs  during this step, execution also halts with an error message.

## Problem Analysis
 Programming is a process of problem solving. Different people use different techniques to solve problems. To become a good programmer, we should have good problem solving techniques. Following steps are commonly used for solving the problems: 

 1. Analyze the problem.
 2. Outline the problem requirements.
 3. Design the problem using algorithm, flowcharts etc.
 4. Implement the algorithm using programmming language like python.
 5. Verify that algorithm works.
 6. Maintain the program by using it, modifying it and improving it if problem domain changes.


 #### Steps for analyzing the problem
 - Understand the problem
 - Understand requirement of the problem. Requirements means what input is like, how to interact with the user, what kind of data is needed and what the output should look like.
 - If the problem is complex, we can divide it into small chunks of sub - problems and understand each subproblem's requirements.

 ### Algorithms 

 - Algorithm is step by step procedure to solve the given problem. It is a set of instructions that is followed to solve a particular problem. 

 ##### Example : 
 #### Algorithm to find the sum of two numbers
 1. Take two numbers as input
 2. Add two numbers using + operator 
 3. Display the result
