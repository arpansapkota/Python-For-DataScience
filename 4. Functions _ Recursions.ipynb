**Return Statement**

Return Statement is used to return result from function. return statement terminates the function execution and takes the control back to the point of function call.

See the program below:

````
a = int(input("Enter first large number: "))
b = int(input("Enter Second number smaller than first number: "))

def arithematic_operation(a,b):
  sum = a + b
  diff = a -b 
  prod = a * b
  div = a / b
  rem = a % b
  divs = a//b

  return sum, diff, prod, div, rem, divs

sum, diff, prod, div, rem, divs = arithematic_operation(a, b)
print("The Sum of two number is", sum)
print("The Difference of two number is", diff)
print("The Product of two number is", prod)
print("The Division of two number is", div)
print("The Reminder of two number is", rem)
print("The Quotient of two number is", divs)
````

**Variable Function Arguments**

Python provides multiple ways to define variable arguments in function.

**1. Default Argument**

> Default argument value is defined in function defination itself. If no any value is passed for argument, parameter will have default value.




````
def greet(name, msg = "Good morning!"):
   """
   This function greets to the person with the provided message.

   If message is not provided, it defaults to "Good morning!"
   """

   print("Hello",name + ', ' + msg)

greet("Rajesh")
greet("Mohan","How do you do?")
````

*****Note: Any number of arguments in a function can have a default value. But once we have a default argument, all the arguments to its right must also have default values i.e. non-default argument cannot follow default argument***

#Recursion

Recursion is the functionality through which function call itself. Recursion is derived from recur which basically means same event occuring again and again.

Recursion would repeat indefinitely if not stopped by some condition. This condition is known as base condition. A base condition is must in every recursive programs otherwise it will continue to execute forever like an infinite loop.

1. Recursive function is called externally for the first time.
2. Recursive condition known as base condition should be met to continue recursion or to exit from recorsion. 

Very common technique of recursion usages are
- Fibonacci series
- Factorial
- Binary Search
- Merge Sort etc.

See the example below:

````
#Recursive factorial implementation
def fact(n, i):
    if n == 0:
        return 1
    else:
        print("Recursion", i,"time with value of n as", n)
        return n * fact(n-1,i+1)

i = 1
print(fact(7, i))
````

````
#Fibonacci Series
def fib(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fib(n-1) + fib(n-2)

print(fib(11))
````

#Lambda Function

Lambda function are the anonymous function in python without any name. If functions are required for short span of time for very specific task only, lambda function are handy.

Normal functions are defined using the def keyword, in Python anonymous functions are defined using the lambda keyword.

Syntax:
```
lambda arguments: expression
```
Lambda functions can have any number of arguments but only one expression. The expression is evaluated and returned. Lambda functions can be used wherever function objects are required.



````
#Using lambda functions
prod = lambda x: x * 3

a=[[3, 4]]
print(prod(a))
````

````
# Program to filter out only the even items from a list
number_list = range(100)

#Get the even list using lambda function
even_list = list(filter(lambda x: (x%2 == 0) , number_list))
print(even_list)
````

# Local Variables

Any binding operation creates a local variable. Examples are:
  
(1) parameters of a function; 

(2) assignment to a variable in a function; 

(3) the import statement; etc.

Inside a function, we must use global when we want to set
the value of a global variable.

````
# Example of local variable
def sum(a, b):
  '''
  Calculating the sum of two variables
  '''
  s = a + b;
  return s;

print(sum(2,3))
print(s)
````

Let's see the global variables:



````
# Example of local variable
def sum(a, b):
  '''
  Calculating the sum of two variables
  '''
  global s 
  s = a + b;
  return s;

print(sum(2,3))
print(s)
````
